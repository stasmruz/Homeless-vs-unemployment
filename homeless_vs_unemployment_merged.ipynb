```python
#import dependency
import pandas as pd
```

```python
# reading two csv files
homeless_pit_df = pd.read_csv('Resources/2020-PIT-Estimates-by-state_cleaned.csv')
unemployment_df = pd.read_csv('Resources/unemployment_data_us_state_cleaned.csv')
  
# using merge function by setting how='inner'
merge = pd.merge(homeless_pit_df, unemployment_df, 
                   on='State', 
                   how='inner')
  
# displaying result
merge.head(55)
```

```python
#drop the imported indexes
merge.drop(merge.columns[0], inplace=True, axis=1)
merge.drop(merge.columns[8], inplace=True, axis=1)
merge.head(4)
```

```python
#reorder columns 
merged_data = merge[["State", "Unemployment_Rate","Unemployment_Rank", "Under_18", "Age_18_to_24", "Age_24_and_over",
                     "Female", "Male", "Transgender", "Overall_Homeless"]]
merged_data.head(60)
```

```python
#check data types
merged_data.dtypes
```

```python
# remove all commas in order to convert object to int
df = merged_data.replace(',','', regex=True)
#  convert the some of the objects to int
df .loc[:, "Under_18"] = df .loc[:, "Under_18"].astype("int")
df .loc[:, "Age_18_to_24"] = df .loc[:, "Age_18_to_24"].astype("int")
df .loc[:, "Age_24_and_over"] = df .loc[:, "Age_24_and_over"].astype("int")
df .loc[:, "Female"] = df .loc[:, "Female"].astype("int")
df .loc[:, "Male"] = df .loc[:, "Male"].astype("int")
df .loc[:, "Transgender"] = df .loc[:, "Transgender"].astype("int")
df .loc[:, "Overall_Homeless"] = df .loc[:, "Overall_Homeless"].astype("int")
#check data type
df.dtypes
```

```python
#export cleaned data to csv format
df.to_csv(r'Resources\homeless_vs_unemployment_merged.csv', index = True)
```
