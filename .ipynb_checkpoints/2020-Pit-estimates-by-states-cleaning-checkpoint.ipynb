```python
# Dependencies
import numpy as np
import pandas as pd
```

```python
# import the 2020 pit by state data
pit_data = pd.read_csv("Resources/2020-PIT-Estimates-by-state.csv")
# identify any null values
pit_data.count()
# identify the data type
print(pit_data.isna().sum())
```

```python
# drop any null valules - lets not drop the null values
pit_data = pit_data.dropna(how="any")
print(pit_data.count())
```

```python
# identify the data type
pit_data.dtypes
```

```python
pit_data.head(5)
```

```python
#remove unnecessary columns
pit_data.drop(pit_data.iloc[:, 9:], inplace=True, axis=1)
pit_data.drop(pit_data.columns[1], inplace=True, axis=1)
pit_data.head(59)
```

```python
#rename columns
pit_data1= pit_data.rename(columns = {"Overall Homeless - Under 18, 2020": "Under_18",
                                       "Overall Homeless, 2020": "Overall_Homeless",
                                      "Overall Homeless - Age 18 to 24, 2020": "Age_18_to_24",
                                      "Overall Homeless - Over 24, 2020": "Age_24_and_over",
                                      "Overall Homeless - Female, 2020": "Female",
                                      "Overall Homeless - Male, 2020": "Male",
                                      "Overall Homeless - Transgender, 2020": "Transgender"})
pit_data1.head(55)
```

```python
#delete rows (american territories)
pit_data1 = pit_data1.drop(labels=3, axis=0)
pit_data1 = pit_data1.drop(labels=8, axis=0)
pit_data1 = pit_data1.drop(labels=12, axis=0)
pit_data1 = pit_data1.drop(labels=27, axis=0)
pit_data1 = pit_data1.drop(labels=42, axis=0)
pit_data1 = pit_data1.drop(labels=50, axis=0)

pit_data1.head(51)
```
