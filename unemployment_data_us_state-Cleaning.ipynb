```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#read table from a link as HTML
unemployment = pd.read_html("https://www.bls.gov/lau/lastrk20.htm")
unemployment[0]
```

```python
#Save unemployment into a data_df
data_df = unemployment[0]
```

```python
# drop NaN 
data_df = data_df.dropna(how="any")
print(data_df.count())
```

```python
data_df.head(55)
```

```python
# delete a single row by index value 0 and Distric of columbia
data = data_df.drop(labels=[54,36], axis=0)
```

```python
# delete column Rank
data.drop(data.columns[2], inplace=True, axis=1)
```

```python
#rename collumn 2020rate to unemployment rate
data.rename(columns={'2020rate':'Unemployment_Rate'},inplace=True)
#reorder columns 
data = data[["State", "Unemployment_Rate"]]
```

```python
data.head(60)
```

```python
#export cleaned data to csv format
data.to_csv(r'Resources\unemployment_data_us_state_cleaned.csv', index = True)
```
